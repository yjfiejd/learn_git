```python
import networkx as nx
```

#### 1） Draw No.2 graph -> all connect graph nodes

```python
# 1) 需要获得right_input ->  画图
right_input_by_edges =[[(57, 58), (1, 57)], [(5, 4), (5, 3)], [(7, 6), (63, 59), (8, 7), (62, 8), (62, 63)], [(67, 66), (67, 10), (10, 64), (66, 65)], [(70, 69)], [(75, 73), (75, 74), (18, 13), (18, 71), (18, 68), (76, 18), (76, 75)], [(77, 82), (77, 78), (83, 77), (83, 25), (83, 26), (21, 20), (78, 21)], [(80, 81)], [(96, 88), (96, 90), (87, 84), (91, 96), (92, 96), (93, 96), (88, 87), (90, 89)], [(86, 85)], [(95, 94), (34, 94)], [(102, 45), (102, 97), (43, 101), (43, 44), (101, 42), (101, 97), (97, 98), (97, 99), (97, 100)], [(106, 105)], [(108, 104), (109, 107), (109, 49), (109, 48), (109, 108)], [(53, 111), (111, 110)]]
right_input_by_edges
```

```python
G_dir = nx.DiGraph()
```

```python
# 2) 边的形式append集合  -> 转为extend格式，可以输入Digraph()中
def change_right_input_edges_form(right_input_by_edges):
    right_input_list = []
    for i in right_input_by_edges:
    #     print(i)
        result = []
        for j in i:
    #         print(j)
            result.append(j)
        right_input_list.extend(result) # 注意使用extend，可以破除这些
    return right_input_list
```

```python
# 【extend 边的集合】
right_input_list = change_right_input_edges_form(right_input_by_edges)  # 可以作为边加入有向图了
right_input_list
```

```python
# 3) 加入有向图中
G_dir.add_edges_from(right_input_list)
```

```python
## 4) 画出有向图 (这是整段话的子图)
import matplotlib.pyplot as plt
plt.figure() 
nx.draw_networkx(G_dir, with_labels=True,)

# font_size=10, node_size=200, arrowsize=8
```

```python
# 5）画出有向图 (这是某句话的子图)， 把每一个句子的图表达出来
for i in right_input_by_edges:
    print(i)
    G = nx.DiGraph()
    G.add_edges_from(i)
    plt.figure() 
    nx.draw_networkx(G, with_labels=True)
    plt.show()
    
```
