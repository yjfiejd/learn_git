```python
# 方法一：找到对应位置的最大值，返回
def pick_the_largest(a, b):
    result = []
    list_length = len(a)
    for i in range(list_length):
        result.append(max(a[i], b[i]))
    return result
```

```python
pick_the_largest(a, b)
```

```python
# 方法二：
result = zip(a, b)
print(result) 
# 注意这个zip在不同版本的不同反应了python的一个演变：大部分返回list的函数不在返回list，而是返回一个支持遍历的对象，比如map、fiter之类的
# 加一个list就行了
print(list(result))
```

```python
# 使用lambda
myfunc = lambda x, y : x + y
aa = myfunc(2, 5)
print(aa)
```

```python
myfunc_2 = lambda x, y: max(x, y)
aaa = myfunc_2(2, 5)
print(aaa)
```

```python
test = lambda pair: zip(a,b)
print(test)
```

```python
#### map()是 Python 内置的高阶函数，它接收一个函数 f 和一个 list，并通过把函数 f 依次作用在 list 的每个元素上，得到一个新的 list 并返回
```

```python
# 最终的表达式

test = map(lambda one_list: max(one_list), zip(a, b))
print(test)
print(list(test))
```
